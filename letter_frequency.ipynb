```python
import pandas as pd
from random import choice
```

```python
letter_frequency = pd.read_csv('data/letter_breakdown.csv')
```

```python
alphabet = ['a','b','c','d','e','f','g',
            'h','i','j','k','l','m','n',
            'o','p','q','r','s','t','u',
            'v','w','x','y','z']

columns = [letter + "_max" for letter in alphabet] 
```

```python
ranking = letter_frequency[columns].sum().sort_values(ascending=False)
top_5_letters = ranking[:5].index
```

```python
ranked_top_words = letter_frequency[top_5_letters].sum(axis=1).sort_values(ascending=False)
```

```python
top_rated_words = ranked_top_words[ranked_top_words==ranked_top_words.max()].index
```

```python
chosen_index = choice(top_rated_words)
chosen_word = letter_frequency.at[chosen_index,'solution_word']
print(chosen_word)
```

```python
for i, letter in enumerate(chosen_word):
    if letter in chosen_word[i+1:]:
        print("duplicate_letter")
        break
    if i+1 == len(chosen_word):
        print(chosen_word)
    
```
